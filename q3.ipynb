```python
import numpy as np
from sklearn.datasets import load_diabetes
from sklearn.linear_model import LinearRegression
from sklearn.kernel_ridge import KernelRidge
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
x,y = load_diabetes(return_X_y=True)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

```

```python
model = LinearRegression()
model.fit(x_train, y_train)
predictions1 = model.predict(x_test)
err1 = np.mean((y_test-predictions1)**2)
```

```python
kernel1 = KernelRidge(kernel='rbf',alpha=0.01) #guassian
kernel1.fit(x_train, y_train)
predictions2 = kernel1.predict(x_test)
err2 = np.mean((y_test-predictions2)**2)
```

```python
kernel2 = KernelRidge(kernel='poly', alpha=0.01, degree=2) #((x^T)y)^2
kernel2.fit(x_train, y_train)
predictions3 = kernel2.predict(x_test)
err3 = np.mean((y_test-predictions3)**2)
```

```python
print(err1, err2, err3)
```
