```python
import urllib.request
import gzip
import numpy as np
import os
```

```python
def download_file(url, filepath):
    if not os.path.exists(filepath):
        urllib.request.urlretrieve(url, filepath)
def load_images(filepath):
    with gzip.open(filepath, 'rb') as f:
        data = np.frombuffer(f.read(), dtype=np.uint8, offset=16)
    return data.reshape(-1, 784)
def load_labels(filepath):
    with gzip.open(filepath, 'rb') as f:
        data = np.frombuffer(f.read(), dtype=np.uint8, offset=8)
    return data
base_url ="http://fashion-mnist.s3-website.eu-central-1.amazonaws.com/"

download_file(base_url + "train-images-idx3-ubyte.gz", "train-images.gz")
download_file(base_url + "train-labels-idx1-ubyte.gz", "train-labels.gz")
download_file(base_url + "t10k-images-idx3-ubyte.gz", "test-images.gz")
download_file(base_url + "t10k-labels-idx1-ubyte.gz", "test-labels.gz")

x_train = load_images("train-images.gz")
y_train = load_labels("train-labels.gz")
x_test  = load_images("test-images.gz")
y_test  = load_labels("test-labels.gz")


y_train=np.eye(10)[y_train] #one-hot encoding
y_test = np.eye(10)[y_test]
print(x_train.shape)
print(y_train.shape)
```

```python
def softmax(z):
  ez = np.exp(z)
  ez = ez/np.sum(ez, axis=1, keepdims=True)
  return ez



def compute_gradient(X,y, W, b):
  m,n = X.shape
  _,c=W.shape
  dj_dw=np.zeros(W.shape)
  dj_db = np.zeros((1,c))

  lin = np.dot(X,W) + b
  f_wb = softmax(lin)
  err = f_wb - y
  dj_dw = np.dot(X.T, err)
  dj_db = np.sum(err, axis=0,keepdims=True)

  dj_dw= dj_dw/m
  dj_db = dj_db/m
  return dj_dw, dj_db
```

```python
def gradient_descent(X, y, W, b):
  alpha = 0.01
  for epoch in range(500):
      dj_dw,dj_db = compute_gradient(X, y, W, b)
      W = W - alpha * dj_dw
      b = b - alpha * dj_db
  return W, b
```

```python
def predict(X, W, b):
    lin = np.dot(X,W) + b
    probabilities = softmax(lin)
    return np.argmax(probabilities, axis=1)
```

```python
W = np.zeros((784, 10))
b = np.zeros((1, 10))

W, b = gradient_descent(x_train, y_train, W, b)
```

```python
predictions = predict(x_test, W, b)
```

```python
accuracy = np.mean(predictions==y_test)*100
print(accuracy)
```
